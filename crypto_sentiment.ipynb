# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Read your api key environment variable
load_dotenv()
api_key = os.getenv("news_api")

# Set News API Key
newsapi = NewsApiClient(api_key=os.environ["news_api"])
```

```python
# Fetch the Bitcoin news articles
bitcoin_headlines = newsapi.get_everything(
    q="bitcoin",
    language="en",
    page_size=50,
    sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Bitcoin: {bitcoin_headlines['totalResults']}")
```

```python
# Fetch the Ethreum news articles
ethereum_headlines = newsapi.get_everything(
    q="ethereum",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Ethereum: {ethereum_headlines['totalResults']}")
```

```python
# Create the Bitcoin sentiment scores DataFrame
Bitcoin_sentiments = []

for article in bitcoin_headlines['articles']:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        Bitcoin_sentiments.append({
            "compound": compound,
            "negative": neg,
            "neutral": neu,
            "positive": pos,
            "text": text
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
bitcoin_df = pd.DataFrame(Bitcoin_sentiments)

# Reorder DataFrame columns
cols = ["compound", "negative", "neutral","positive", "text"]
bitcoin_df = bitcoin_df[cols]


bitcoin_df.drop_duplicates(subset='compound', keep="first", inplace=True)

bitcoin_df.reset_index(drop=True, inplace=True)

bitcoin_df
```

```python
# Create the ethereum sentiment scores DataFrame
Ethereum_sentiments = []

for article in ethereum_headlines['articles']:
    try:
        text1 = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text1)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        Ethereum_sentiments.append({
            "compound": compound,
            "negative": neg,
            "neutral": neu,
            "positive": pos,
            "text": text1
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
ethereum_df = pd.DataFrame(Ethereum_sentiments)

# Reorder DataFrame columns
cols = ["compound", "negative", "neutral","positive", "text"]
ethereum_df = ethereum_df[cols]

ethereum_df.drop_duplicates(subset='compound', keep="first", inplace=True)

ethereum_df.reset_index(drop=True, inplace=True)

ethereum_df
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
lemmatizer = WordNetLemmatizer()
```

```python
# Expand the default stopwords list if necessary
nltk.download('stopwords')
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""

    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    tokens = [word.lower() for word in lem if word.lower() not in sw]
    
    return tokens

bitcoin_tokens = []
for text in bitcoin_df['text']:
    bitcoin_tokens.append(tokenizer(text))
    
```

```python
# Create a new tokens column for bitcoin
bitcoin_data = pd.DataFrame(
    {"tokens": bitcoin_tokens})

bitcoin_data.head(10)
```

```python
# Merge dataframes
bitcoin_final = bitcoin_df.join(bitcoin_data)

# # Reorder DataFrame columns
cols = ["compound", "negative", "neutral", "positive", "text", "tokens"]
bitcoin_final = bitcoin_final[cols]
```

```python
# Create a new tokens column for ethereum
def tokenizer(text1):
    """Tokenizes text."""

    # Convert the words to lowercase
    regex = re.compile("[^a-zA-Z ]")
    
    # Remove the punctuation
    re_clean = regex.sub('', text1)
    words = word_tokenize(re_clean)
    
    # Remove the stop words
    sw = set(stopwords.words('english'))
    
    
    # Lemmatize Words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
    tokens1 = [word.lower() for word in lem if word.lower() not in sw]
    
    return tokens1

# For loop over each of the text for tokenization
ethereum_tokens = []
for text1 in ethereum_df['text']:
    ethereum_tokens.append(tokenizer(text1))
```

# Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
# Generate the Bitcoin word cloud
def process_text(doc):
    sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem if word.lower() not in sw]
    return ' '.join(output)
```

```python
bitcoin_word = []
for text in bitcoin_df['text']:
    bitcoin_word.append(process_text(text))
    big_string = ' '.join(bitcoin_word)
input_words = process_text(big_string)
wc = WordCloud().generate(input_words)
plt.imshow(wc)
```

```python
ethereum_word = []
for text1 in ethereum_df['text']:
    ethereum_word.append(process_text(text1))
    big_string = ' '.join(ethereum_word)
input_words1 = process_text(big_string)
wc1 = WordCloud().generate(input_words1)
plt.imshow(wc1)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
# !python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
bitcoin_ner = ', '.join(bitcoin_word)
# print(bitcoin_ner)
```

```python
# Run the NER processor on all of the text
doc = nlp(bitcoin_ner)
print("NER PROCESSOR FOR BITCOIN")
for ent in doc.ents:
    print(ent.text, ent.label_)
# Add a title to the document
```

```python
# Render the visualization
displacy.render(doc, style='ent')
```

```python
# List all Entities
print([ent.text for ent in doc.ents if ent.label_ == 'ORG'])
```
